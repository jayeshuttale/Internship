```python
# web page autometion

driver = webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
#fetching views

Views = []
view = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in view:
    Views.append(i.text)
```

```python
#fetching uplode_date

uplode_date = []
date = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in date:
    uplode_date.append(i.text)
```

```python
driver.close()
```

```python
# Creating Data Frame

you_tube_df = pd.DataFrame({
    "Rank"           : range(1,31),
    "Name of video"  : Name,
    "Uploder name"   : uploder_name,
    "Views(billions)": Views,
    "Uplode Date"    : uplode_date
})
you_tube_df.set_index("Rank",inplace = True)
you_tube_df
```

```python
#Web driver autometion
driver = webdriver.Chrome()
driver.get("https://github.com/")  ## fatching web page
```

```python
#hovering on list
action = ActionChains(driver)
action.move_to_element(driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')).perform()
```

```python
#fetching tranding repo. page
tranding_repo = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
tranding_repo.click()
t.sleep(3)
```

```python
#extracting data
Title = []
title = driver.find_elements(By.XPATH,'//a[@class="Link"]')
for i in title:
    Title.append(i.text)
```

```python
# fetching data from links
links = []
url = driver.find_elements(By.XPATH,'//a[@class="Link"]')
for i in url:
    links.append(i.get_attribute("href"))
```

```python
# creating empty list
Description = []
Contributors_count = []
Language_used = []

for Url in links:
    driver.get(Url)
    t.sleep(2)

    try:
        description = driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        Description.append(description.text)
    except NoSuchElementException:
        Description.append("Description Not Present")

    try:
        count = driver.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[5]/div/h2/a/span")
        Contributors_count.append(count.text)
    except NoSuchElementException:
        Contributors_count.append('Not present')
        
    try:
        Language = driver.find_element(By.XPATH,'//ul[@class="list-style-none"]')
        Language_used.append(Language.text)
    except NoSuchElementException:
        Language_used.append("-")
```

```python
#cleaning data
import re
new_Language_used = []
for l in Language_used:
    new_Language_used.append(re.sub(r"\n\d+.\d+%|\n"," ",l))
```

```python
driver.close()
```

```python
#creting dataFrame
git_df = pd.DataFrame({
    "Title"             : Title,
    "Description"       : Description,
    "Contributors count": Contributors_count,
    "Language used"     : new_Language_used
})
git_df
```

```python
#webpage autometion
driver = webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
#fetching page
dataset = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/header/nav/ul/li[1]/a')
dataset.click()
```

```python
#procesing page
page = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/label/select')
page.click()
option = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/label/select/option[5]')
option.click()
```

```python
#creting empty list
Dataset_name = [] 
Data_type  = []
Task = []
Attribute_type = [] 
No_of_instances  = []
No_of_attribute = []
Year = []
```

```python
#extracting data from mutiple page
start, end = 0,28
for page1 in range(start,end):
    
    name = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in name:
        Dataset_name.append(i.text)
    
    d_type = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
    for i in d_type:
        Data_type.append(i.text)
        
    task = driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/p')
    for i in task:
        Task.append(i.text)
        
    a_type = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span')
    for i in a_type:
        Attribute_type.append(i.text)
        
    instance = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
    for i in instance:
        No_of_instances.append(i.text)
        
    attribute = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
    for i in attribute:
        No_of_attribute.append(i.text)
        
    svg = driver.find_elements(By.XPATH,"""//*[local-name()='svg' and @class="absolute right-1 top-1/2 -translate-y-1/2 fill-primary transition-transform"]""")
    for i in svg:
        i.click()
        t.sleep(2)
    
    year = driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[3]')
    for i in year:
        Year.append(i.text)

        
    next_button = driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
```

```python
driver.close()
```

```python
#creting dataframe

dataset_df = pd.DataFrame({
    "Dataset name"    : Dataset_name,
    "Data type"       : Data_type,
    "Task"            : Task,
    "Attribute type"  : Attribute_type,
    "No of instances" : No_of_instances,
    "No_of_attribute" : No_of_attribute,
    "Year"            : Year
})
dataset_df
```

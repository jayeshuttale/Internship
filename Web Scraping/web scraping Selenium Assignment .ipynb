```python
# import necessary libraries

import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filter="ignore"
import time
```

```python
# making web page automation

driver = webdriver.Chrome()
driver.get("https://www.shine.com/")

# clicking search button

Click = driver.find_element(By.CLASS_NAME,"iconH-zoom-white")
Click.click()
time.sleep(2)

# adding skill in web page 

skill = driver.find_element(By.CLASS_NAME,"form-control  ")
skill.send_keys("Data Analyst")
time.sleep(2)

#adding location in web page

loc = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
loc.send_keys("Bangalore")
time.sleep(2)

# clicking search button 

search = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
search.click()
time.sleep(2)
```

```python
# creating empty list

job_title = []
company_name = []
Experiance_required = []
job_location = []

# extracting job title

title = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]')
for i in title[0:10]:
    job_title.append(i.text)
    
time.sleep(2)

# extracting companty name

Name = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in Name[0:10]:
    company_name.append(i.text)

time.sleep(2)

# extracting experiance required

exp = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in exp[0:10]:
    Experiance_required.append(i.text)

time.sleep(2)

# extracting job location
import re
job_loc=[]
loc=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in loc[0:10]:
    job_loc.append(i.text)
for j in job_loc:
    l=re.sub("\n\+\d+","",j)
    job_location.append(l)
    
    
job_location
time.sleep(2)
```

```python
#creatind data frame

df_bangalore = pd.DataFrame({
    "Job Title":job_title,
    "Company":company_name,
    "Experiance Required":Experiance_required,
    "Job Location": job_location
})
df_bangalore
```

```python
driver.close()
```

```python
 # import necessary libraries

import pandas as pd
import selenium 
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings 
warnings.filter="ignore"
import time
```

```python
# making web page automation

driver = webdriver.Chrome()
driver.get("http://www.azquotes.com/")
```

```python
#pressing top quote 

search = driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
search.click()
time.sleep(2)
```

```python
#creating empty list

Quote = []
Author = []
Type_Of_Quotes = []
```

```python
# extracting data from mutiple pages
start,end = 0,10

for page in range(start,end):
    
    quote = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote:
        Quote.append(i.text)
        
    author = driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for j in author:
        Author.append(j.text)
        
    type_of_quotes = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for k in type_of_quotes:
        Type_Of_Quotes.append(k.text)
        
    next_button = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    next_button.click()
    time.sleep(3)
```

```python
len(Quote),len(Author),len(Type_Of_Quotes)
```

```python
#creating dataframe
Quote_df = pd.DataFrame({
    "Quote" : Quote,
    "Author": Author,
    "Type Of Quotes" : Type_Of_Quotes
})

Quote_df
```

```python
driver.close()
```

```python
#import necessary libraries

import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filter = "ignore"
import time
```

```python
#making web page automation

driver = webdriver.Chrome()
driver.get("https://www.motor1.com/")

#searching 

search = driver.find_element(By.XPATH,"/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/input")
search.send_keys("50 most expensive cars")

#pressing search button

search_button = driver.find_element(By.XPATH,"/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/button[1]")
search_button.click()
time.sleep(2)
```

```python
#clicking 50 most expensive cars in world

button = driver.find_element(By.XPATH,"/html/body/div[10]/div[9]/div/div[1]/div/div/div[2]/div/div[1]")
button.click()
time.sleep(2)
```

```python
#creating empty list

Car_name = []
price = []
```

```python
# extracting data of car name

car = driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in car[0:50]:
    Car_name.append(i.text)
Car_name
time.sleep(2)
```

```python
# extracting data of car price

import re

rate = driver.find_elements(By.XPATH,'//div[@class="postBody description e-content"]/p/strong')
for i in rate:
    price.append(re.sub("\w+:\s|\(est.\)","",i.text))
price
time.sleep(2)
```

```python
#creating data frame

Cars_df = pd.DataFrame({
    "Cars Name":Car_name,
    "Price"    :price
})

Cars_df
```
